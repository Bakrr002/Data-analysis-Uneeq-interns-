```python
print(df.duplicated().sum())
print(df.isnull().any())
print(df.isnull().sum())
```

```python
cols = ['BDMG_0', 'BDMG_1', 'BDMG_2', 'BDMG_3', 'HDMG_0', 'HDMG_1', 'HDMG_2', 'HDMG_3']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python

df['BDMG_0']=df['BDMG_0'].fillna(df['BDMG_0'].mean())
df['BDMG_1']=df['BDMG_1'].fillna(df['BDMG_1'].mean())
df['BDMG_2']=df['BDMG_2'].fillna(df['BDMG_2'].mean())
df['BDMG_3']=df['BDMG_3'].fillna(df['BDMG_3'].mean())
df['HDMG_0']=df['HDMG_0'].fillna(df['HDMG_0'].mean())
df['HDMG_1']=df['HDMG_1'].fillna(df['HDMG_1'].mean())
df['HDMG_2']=df['HDMG_2'].fillna(df['HDMG_2'].mean())
df['HDMG_3']=df['HDMG_3'].fillna(df['HDMG_3'].mean())
```

```python
df.tail()
```

```python
df.drop_duplicates()
```

```python
plt.figure(figsize=(8,6))
df['Weapon Type'].hist(bins=20,color='#ff7f0e',edgecolor='#d62728',grid=False)
plt.xlabel('Weapon Type')
plt.xticks(rotation=45)
plt.ylabel('Freq')
plt.title('Nums of Guns')
plt.show()
```
